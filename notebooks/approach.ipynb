```python
apr_means[AVG_DISTANCE TO PIN IN FEET]
```

```python
sns.set_theme(style="darkgrid")
plt.figure(figsize=(15, 10))
ax = sns.lineplot(data=apr_means, x="YARDS OUT", y="AVG_DISTANCE TO PIN IN FEET", marker='o', palette="crest")

for i in range(len(apr_means)):
    plt.annotate(f'{apr_means["AVG_DISTANCE TO PIN IN FEET"][i]:.0f}', (apr_means["YARDS OUT"][i], apr_means["AVG_DISTANCE TO PIN IN FEET"][i]), 
                 textcoords="offset points", xytext=(0,25), ha='left')

plt.xticks(rotation=45)
plt.grid("light")
plt.xlabel('YARDS')
plt.ylabel('AVG_DISTANCE TO PIN IN FEET')
plt.title('PGA TOUR AVG DISTANCE TO PIN IN FEET 2022-2023')
plt.show();
```
