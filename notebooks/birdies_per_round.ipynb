```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
pga_birdies_per_round = pd.read_csv(r"C:\Users\Jeremy\Documents\NSS DA 2024\captsone\data\comparable\birdies per round\PGA birdie avg per round.csv")
```

```python
pga_birdies_per_round.head()
```

```python
pga_birdies_per_round['AVG'].describe()
```

```python
pga_birdies_per_round = pga_birdies_per_round.drop(columns = ['MOVEMENT', 'PLAYER_ID'])
pga_birdies_per_round.head()
```

```python
sns.set_theme(style="darkgrid")
plt.figure(figsize=(15, 10))
sns.boxplot(data = pga_birdies_per_round, x = 'AVG', palette="crest")
plt.xlabel("BIRDIES PER ROUND")
plt.title("PGA TOUR BIRDIES PER ROUND AVG 2022-2023");
```

```python
am_birdies_per_round = pd.read_csv(r"C:\Users\Jeremy\Documents\NSS DA 2024\captsone\data\comparable\birdies per round\am_birdies _per_round.csv")
```

```python
sns.set_theme(style="darkgrid")
plt.figure(figsize=(15, 10))
ax = sns.barplot(am_birdies_per_round, x="Handicap Level", y="Birdies Per Round", palette="crest")
plt.xticks(rotation=30)
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f', padding=0.5)
plt.xlabel("HANDICAP")
plt.title("AMATEUR BIRDIES PER ROUND AVG")
plt.show();
```
