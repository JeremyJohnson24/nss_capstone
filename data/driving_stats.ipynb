```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
pga_driving_stats = pd.read_csv(r"C:\Users\Jeremy\Documents\NSS DA 2024\captsone\data\comparable\driving\pga_driving_2022_2023.csv")
```

```python
pga_driving_stats
```

```python
pga_driving_stats = pga_driving_stats.drop(columns = ['MOVEMENT', 'PLAYER_ID'])
pga_driving_stats.head()
```

```python
sns.set_theme(style="darkgrid")
plt.figure(figsize=(15, 10))
sns.boxplot(data = pga_driving_stats, x = 'AVG', palette="crest")
plt.xlabel("YARDS")
plt.title("2022-2023 PGA TOUR DRIVE DISTANCE AVERAGES");
```

```python
plt.figure(figsize=(20, 15))
ax = sns.barplot(pga_driving_stats.head(25), x="PLAYER", y="AVG",palette="crest")
plt.xticks(rotation=90)
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f', padding=0.1, rotation=45)
plt.grid()
plt.show();
```

```python
plt.figure(figsize=(20, 15))
ax = sns.barplot(pga_driving_stats.tail(25), x="PLAYER", y="AVG",palette="crest")
plt.xticks(rotation=90)
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f', padding=0.5, rotation=45)
plt.grid()
plt.show();
```

```python
pga_driving_stats_describe = pga_driving_stats['AVG'].describe().reset_index()
pga_driving_stats_describe
```
