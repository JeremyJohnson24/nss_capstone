```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
am_ppr = pd.read_csv(r"C:\Users\Jeremy\Documents\NSS DA 2024\captsone\data\comparable\putting\AM putts per round.csv")
```

```python
am_ppr.head(6)
```

```python
am_ppr['Average Putts per Round'].head(6).describe()
```

```python
sns.set_theme(style="darkgrid")
plt.figure(figsize=(15, 10))
ax = sns.barplot(am_ppr.head(6), x="Handicap", y="Average Putts per Round", palette="crest")
plt.xticks(rotation=30)
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f', padding=0.5)
plt.grid("light")
plt.xlabel("HANDICAP")
plt.ylabel("PUTTS")
plt.title("AMATEURS AVG PUTTS PER ROUND")
plt.show();
```

```python
sns.set_theme(style="darkgrid")
plt.figure(figsize=(15, 10))
sns.boxplot(am_ppr.head(6), x="Average Putts per Round", palette="BuGn")
plt.xlabel("PUTTS")
plt.title("AMATEUR PUTTS PER ROUND");
```
