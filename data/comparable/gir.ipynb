```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
am_gir = pd.read_csv(r"C:\Users\Jeremy\Documents\NSS DA 2024\captsone\data\comparable\GIR\AM GIR all distances.csv")
```

```python
am_gir
```

```python
am_gir['GIR Percentage'] = am_gir['GIR Percentage'].str[:-1]
am_gir['GIR Percentage'] = am_gir['GIR Percentage'].astype(float)
am_gir
```

```python
sns.set_theme(style="darkgrid")
plt.figure(figsize=(15, 10))
ax = sns.barplot(am_gir, x="Handicap", y="GIR Percentage", palette="crest")
plt.xticks(rotation=30)
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f', padding=0.5)
plt.grid("light")
plt.xlabel("HANDICAP")
plt.ylabel("PERCENT")
plt.title("AMATEURS AVG GREENS IN REGULATION PERCENTAGE")
plt.show();
```
