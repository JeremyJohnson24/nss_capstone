```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
am_par_3_scores = pd.read_csv(r"C:\Users\Jeremy\Documents\NSS DA 2024\captsone\data\comparable\par3_4_5 scoring\am-pa3_scores.csv")
```

```python
am_par_3_scores
```

```python
sns.set_theme(style="darkgrid")
plt.figure(figsize=(15, 10))
ax = sns.barplot(am_par_3_scores, x="Handicap", y="Average Score", palette="crest")
plt.xticks(rotation=30)
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f', padding=0.5)
plt.grid("light")
plt.xlabel("HANDICAP")
plt.ylabel("AVERAGE SCORE")
plt.title("AMATEURS AVG SCORE ON PAR 3 HOLES")
plt.show();
```
